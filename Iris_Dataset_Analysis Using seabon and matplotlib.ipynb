# Line Plot:

*   Plot a line graph showing the sepal length of the three species of Iris flowers. Use different colors to represent each species.


```python
# group species data
grouped_data=df.groupby('Species')['SepalLengthCm']
```

```python
versicolor = grouped_data.get_group('Iris-versicolor')
virginica = grouped_data.get_group('Iris-virginica')
setosa= grouped_data.get_group('Iris-setosa')
```

```python
# plot lineolot

plt.figure(figsize=(8,6))
# x-axis lenth
x=range(1,len(versicolor)+1)
# Plot sepal length for each species
plt.plot(x, versicolor.values, label='Iris-versicolor')
plt.plot(x, virginica.values, label='Iris-virginica')
plt.plot(x, setosa.values, label='Iris-setosa')
# Set labels and title
plt.xlabel('Index', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.title('Sepal Length of Iris Species', fontsize=14)
plt.legend()
plt.grid(True)
plt.savefig('sepal_length_plot.png')  # Save the graph as an image
```

# Scatter Plot:

*   Create a scatter plot showing the relationship between the petal length and petal width. Use different colors to represent each species.




```python
# Scatter plot
sns.scatterplot(data = df,x ='PetalWidthCm',y='PetalLengthCm',hue='Species', palette='magma')
plt.grid(True)
plt.title('Relation between Petalwidth & Petallengt',fontsize=16)
plt.savefig('Scatterplot of petalwidth and petalLength .png')  # Save the graph as an image
# plot show
plt.show()
```

# Histogram:

*   Plot a histogram showing the distribution of the petal widths. Use different colors to represent each species and display a legend.




```python
# histrogram
sns.histplot(data=df,x='PetalWidthCm',hue ='Species',bins =40,palette='magma')
# add title
plt.title('Histrogram of Petal Width')
plt.savefig('Histrogram.png')  # Save the graph as an image
```

# Bar Plot:

*    Create a bar plot showing the average sepal width for each species.




```python
# group species and find average of sepal width
ave = df.groupby('Species')['PetalWidthCm'].mean().reset_index()
# plot bar plot
sns.barplot(data=ave,x='Species',y='PetalWidthCm',hue ='Species',palette='magma')
# add title and labels
plt.title('Average of sepalwidth for ecah species',fontsize = 16)
plt.xlabel('Species',fontsize = 13)
plt.ylabel('Average of sepalwidth',fontsize = 13)
# rotate x axis
plt.xticks(rotation = 45)
plt.savefig('Ave sepalwidth barplot.png')  # Save the graph as an image
# Show plot
plt.show()
```

# Pie Chart:


*  Assume that the dataset represents the entire population of Iris flowers. Create a pie chart showing the proportion of each species in the population.



```python
# plot pie chart
df['Species'].value_counts().plot(kind='pie',autopct="%.1f%%")
# add title
plt.title('Porportion of each species')
plt.savefig('pie plot of each species.png')  # Save the graph as an image
```

# Box Plot:

*   Create a box plot showing the distribution of sepal length for each species.




```python
# Plot box plot
sns.boxplot(data=df, x='Species', y='PetalLengthCm',palette ='rocket')

# Add title
plt.title('Box plot of each species')
plt.savefig('boplot of each species.png')  # Save the graph as an image
```

# Subplots:


*   Combine the scatter plot and the histogram into one figure by using subplots.



```python
# set figure size
plt.figure(figsize=(10,6))
# subplot to combine 2 plot
plt.subplot(121)
# plot boxplot
sns.boxplot(data=df, x='Species', y='PetalLengthCm',palette='magma')
# add title
plt.title('Boxplot')
# subplot to combine 2 plot
plt.subplot(122)
# plot scatterplot
sns.scatterplot(data = df,x ='PetalWidthCm',y='PetalLengthCm',hue='Species',palette='magma')
# add title
plt.title('Scatterplot')
plt.savefig('subplot.png')  # Save the graph as an image
# show plot
plt.show()
```
